```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters are given data for each problem

def my_general_lorenz(sigma, beta, rho, initial_condition, start_time, end_time, h):
    # Define the Lorenz equations
    def lorenz_eqns(time, current_lorenz_solution, sigma, beta, rho):
        x, y, z = current_lorenz_solution
        x_derivative = sigma * (y - x)
        y_derivative = x * (rho - z) - y
        z_derivative = x * y - beta * z

        return [x_derivative, y_derivative, z_derivative]

    # Set up the time array and solution array
    time = np.arange(start_time, end_time + h, h)
    current_lorenz_solution = np.zeros((len(time), 3))
    current_lorenz_solution[0] = initial_condition

    # Solve the Lorenz equations using RK4 scheme
    for i in range(1, len(time)):
        k1 = h * np.array(lorenz_eqns(time[i-1], current_lorenz_solution[i-1], sigma, beta, rho))
        k2 = h * np.array(lorenz_eqns(time[i-1] + h/2, current_lorenz_solution[i-1] + k1/2, sigma, beta, rho))
        k3 = h * np.array(lorenz_eqns(time[i-1] + h/2, current_lorenz_solution[i-1] + k2/2, sigma, beta, rho))
        k4 = h * np.array(lorenz_eqns(time[i-1] + h, current_lorenz_solution[i-1] + k3, sigma, beta, rho))
        current_lorenz_solution[i] = current_lorenz_solution[i-1] + (k1 + 2*k2 + 2*k3 + k4)/6
    return current_lorenz_solution
first_solution = my_general_lorenz(10, 8/3, 28, [-8, 8, 27], 0, 8, 0.01)
# Plot the numerical solution
fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot(111, projection='3d')
ax.plot(first_solution[:, 0], first_solution[:, 1], first_solution[:, 2], color='red')
ax.set_xlabel('x(t)')
ax.set_ylabel('y(t)')
ax.set_zlabel('z(t)')
ax.set_title('Figure 1: RK4 Scheme h = 0.01')
# You can change the viewing angle of the diagram with this code
# ax.view_init(azim=80, elev=100)
plt.show()

second_solution = my_general_lorenz(10, 8/3, 28, [-8, 8, 27], 0, 8, 0.005)
# Plot the numerical solution
fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot(111, projection='3d')
ax.plot(first_solution[:, 0], first_solution[:, 1], first_solution[:, 2], color='red')
ax.plot(second_solution[:, 0], second_solution[:, 1], second_solution[:, 2], color='blue')
ax.set_xlabel('x(t)')
ax.set_ylabel('y(t)')
ax.set_zlabel('z(t)')

ax.set_title('Figure 2: RK4 Scheme h = 0.05 and h = 0.01')
# You can change the viewing angle of the diagram with this code
# ax.view_init(azim=80, elev=100)
plt.show()

import matplotlib.pyplot as plt

# Get x(t) and y(t) from the first solution
x1 = first_solution[:, 0]
y1 = first_solution[:, 1]

# Get x(t) and y(t) from the second solution
x2 = second_solution[:, 0]
y2 = second_solution[:, 1]

# Plot x(t) vs y(t) for the first solution (red)
plt.figure(figsize=(8, 6))
plt.plot(x1, y1, color='red', label='h=0.01')

# Plot x(t) vs y(t) for the second solution (blue)
plt.plot(x2, y2, color='blue', label='h=0.005')

plt.xlabel('x(t)')
plt.ylabel('y(t)')
plt.title('PART B: Figure 2: 2D plot of x(t) vs y(t) for two solutions')
plt.legend()
plt.show()



third_solution = my_general_lorenz(10, 8/3, 14, [-8, 8, 27], 0, 8, 0.01)
# Plot the numerical solution
fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot(111, projection='3d')
ax.plot(third_solution[:, 0], third_solution[:, 1], third_solution[:, 2], color='red')
ax.set_xlabel('x(t)')
ax.set_ylabel('y(t)')
ax.set_zlabel('z(t)')

ax.set_title('Figure 3: RK4 Scheme sigma = 10, beta = 8/3, rho = 14')
# You can change the viewing angle of the diagram with this code
# ax.view_init(azim=80, elev=100)
plt.show()

fourth_solution = my_general_lorenz(10, 8/3, 28, [0, 0, 10], 0, 8, 0.01)
# Plot the numerical solution
fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot(111, projection='3d')
ax.plot(fourth_solution[:, 0], fourth_solution[:, 1], fourth_solution[:, 2], color='red')
ax.set_xlabel('x(t)')
ax.set_ylabel('y(t)')
ax.set_zlabel('z(t)')

ax.set_title('Figure 4: RK4 Scheme initial condition = [0, 0, 10]')
# You can change the viewing angle of the diagram with this code
# ax.view_init(azim=80, elev=100)
plt.show()

fifth_solution = my_general_lorenz(10, 8/3, 28, [5, 5, 5], 0, 15, 0.01)
sixth_solution = my_general_lorenz(10, 8/3, 28, [5.0091, 4.9997, 5.0060], 0, 15, 0.01)
# Plot the numerical solution
fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot(111, projection='3d')
ax.plot(fifth_solution[:, 0], fifth_solution[:, 1], fifth_solution[:, 2], color='red')
ax.plot(sixth_solution[:, 0], sixth_solution[:, 1], sixth_solution[:, 2], color='blue')
ax.set_xlabel('x(t)')
ax.set_ylabel('y(t)')
ax.set_zlabel('z(t)')

ax.set_title('Figure 5: RK4 Scheme w/ 2 initial conditions')
# You can change the viewing angle of the diagram with this code
# ax.view_init(azim=80, elev=100)
plt.show()


import matplotlib.pyplot as plt
import numpy as np

# Extract the time values and z(t) values for the fifth solution
t1 = np.arange(0, 15.01, 0.01)
z1 = fifth_solution[:, 0]

# Extract the time values and z(t) values for the sixth solution
t2 = np.arange(0, 15.01, 0.01)  # <-- adjust the endpoint to match t1
z2 = sixth_solution[:, 0]

# Plot the values of z(t) against t for the fifth solution (red)
plt.figure(figsize=(8, 6))
plt.plot(t1, z1, color='red', label='Initial Condition [5,5,5]')

# Plot the values of z(t) against t for the sixth solution (blue)
plt.plot(t2, z2, color='blue', label='Initial Condition [5.0091,4.9997,5.0060]')

plt.xlabel('t')
plt.ylabel('z(t)')
plt.title('Figure: Values of z(t) against t for two initial conditions')
plt.legend()
plt.show()
```
